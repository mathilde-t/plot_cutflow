```python
import matplotlib.pyplot as plt
import mplhep
import pickle
import numpy as np
import sys
```

```python
def get_hist_values(pickle_file, category):
    file_obj = open(path22, 'rb')
    data = pickle.load(file_obj)
    cuts = list(data.axes['step']) # or hist.axes['step']
    values = []
    sums = []
    for cut_name in list(cuts): # hist.axes[1]
        values.append(data[category,:,f'{cut_name}',:,:].values())
        sums.append(data[category,:,f'{cut_name}',:,:].sum())
    return cuts, values, sums

def create_cutflow_histogram(cuts, data, xlabel="Selections", ylabel="Selection efficiency", title="", log=False, rel=False, save_path=None):
    """
    Create a cutflow histogram using Matplotlib with CMS style and save it to a PDF file.

    Parameters:
    - cuts: List of strings representing the names of the cuts / steps.
    - data: List of integers representing the corresponding event counts for each cut.
    - xlabel: Label for the x-axis (default is "Cuts").
    - ylabel: Label for the y-axis (default is "Events").
    - title: Title of the plot (default is "Cutflow Histogram").
    - save_path: Path to save the PDF file. If None, the plot will be displayed but not saved.

    Returns:
    - fig: Matplotlib figure object.
    - ax: Matplotlib axis object.
    """

    # Set CMS style
    plt.style.use(mplhep.style.CMS)
   
    # Create Matplotlib figure and axis
    fig, ax = plt.subplots()

    if log: plt.yscale('log')

    for i, step in enumerate(cuts):
        n_evt = sums[i]['value']
        # Adding label for legend
        #ax.step(step, round(n_evt, 2), color='black', where='mid', marker='o', linewidth=2, alpha=0.8, label=f'Cut: {step}')
        #ax.stairs(
        #    values=round(n_evt, 2),  # heights of the steps
        #    edges=step,               # x positions for the steps
        #    orientation='vertical',   # orientation
        #    fill=False,                # No filling of area
        #    color='black',             # Color of the steps
        #    linewidth=1.2,             # Line width
        #    alpha=0.8                  # Transparency
        #)
        ax.step(step, round(n_evt, 2), color='black', where='mid',marker='o', linewidth=1.2, alpha=0.8)#, label=r"Tau_2022C Data-tobeset")

        
        the_txt = f'{n_evt:.4f}' if rel else f'{n_evt:.0f}'
        ax.annotate(the_txt, (step, n_evt), textcoords="offset points", xytext=(0,-20), ha='center', fontsize=10)

        if log:
            if rel:
                ax.set_ylim((1 * 10**-6, 2 * 10**0))
            else:
                pow_nevt = int(np.log10(n_evt) + 1)
                ax.set_ylim((10**(2), 10**pow_nevt))

    # Set labels and title
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.set_xticks(range(len(cuts)))
    ax.set_xticklabels(cuts, rotation=45, ha='right')

    # Add legend
    #ax.legend()  # Now the legend should appear because labels are added
    ax.grid(True, which='both', axis='y')
    ax.grid(True, which='major', axis='x')
    
    label_options = {
        "wip": "Work in progress",
        "pre": "Preliminary",
        "pw": "Private work",
        "sim": "Simulation",
        "simwip": "Simulation work in progress",
        "simpre": "Simulation preliminary",
        "simpw": "Simulation private work",
        "od": "OpenData",
        "odwip": "OpenData work in progress",
        "odpw": "OpenData private work",
        "public": "",
    }
    cms_label_kwargs = {
        "ax": ax,
        "llabel": label_options.get("simwip"),
        "fontsize": 22,
        "data": True,
        'rlabel': cuts
    }
    mplhep.cms.label(**cms_label_kwargs)
    plt.tight_layout(pad=3.0)  # Increase padding to ensure no overlap
    
     # Save to PDF if save_path is provided
    if save_path:
        fig.savefig(save_path, bbox_inches='tight')
        print(f"Plot saved to {save_path}")
    else:
        # Show the plot if save_path is not provided
        plt.show()

    return fig, ax





path22 = '/Users/mathildewitt/workspace/ma/stras/plot_cutflow/INPUT/cutflow_hist__cf_npvs.pickle'
cat_mutau = 2
cat_incl = 1 # 0 = ?

cuts, values, sums = get_hist_values(path22, cat_mutau)

create_cutflow_histogram(cuts, 
                         data={"2022 preEE": values, "sums": sums},
                         save_path="/Users/mathildewitt/workspace/ma/stras/plot_cutflow/OUTPUT/cutflow_histogram_npvs.png",
                         ylabel="Event Number",
                         log=False,
                         rel=False)
```

```python
import mplhep as hep
hep.style.use("CMS")
#hep.cms.text("Simulation", loc=0)
hep.cms.text("Work in Progress", loc=0)
# Assuming you already have 'cuts' and 'sum_values' from the previous calculation
# Assuming cuts is a list of cut names and sum_values is calculated as in your example

# Number of bins in the 'n' axis (we have 11 bins from 0 to 10)
n_bins = len(data.axes[2].edges) - 1  # 11 bins for 'n' axis, edges array has one extra element

# List to store the summed values for each cut
sum_values = []

# Loop over the third axis (n)
for n in range(n_bins):
    summed_values = np.sum(data[category, :, n, :, :].values())  # Adjust as needed for your data slicing
    sum_values.append(summed_values)

# Convert the list to a numpy array for easy plotting
sum_values = np.array(sum_values)

# Create the stairs plot
plt.step(np.arange(len(cuts)), sum_values, where='mid', linestyle='-', color='blue', alpha=0.7)

# Set labels for the x-axis with the cuts names
plt.xticks(np.arange(len(cuts)), cuts, rotation=45, ha='right')  # Use cuts as labels, rotate for readability

# Set labels and title
plt.xlabel('Cuts')
plt.ylabel('Selection Yield')
plt.title(r"$\mu \tau_h$", loc='right', pad=20)

# Add the sum above each data point
for i, sum_value in enumerate(sum_values):
    plt.text(i, sum_value, f'{sum_value:.1f}', ha='center', va='bottom', fontsize=10)  # Display the sum above the point

# Add grid for better readability
plt.grid(True)

# Show the plot
plt.tight_layout()  # Ensure labels and text don't overlap
plt.savefig("/Users/mathildewitt/workspace/ma/stras/plot_cutflow/OUTPUT/cutflow_histogram_npvs.pdf", bbox_inches='tight')
plt.show()
```
